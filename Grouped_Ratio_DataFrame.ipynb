### For each hour, calculate the ratio of the numbers of patients arrived in each of CTAS 2,3,4,5 to CTAS 1

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('hour_ctas_ratio.xlsx')
df.head(20)
```

```python
df_small=[]
for hour in df['Hour'].unique():
    
    df_ = df[df['Hour']==hour].copy()
    
    for ctas in [1,2,3,4,5]:
        df_[ctas]=df_['Count']/(df['Count'].shift(ctas))
    
    df_['ratio']=df_[[1,2,3,4,5]].ffill(axis=1).iloc[:, -1]
    df_ = df_.drop(columns=[1,2,3,4,5])
    df_['ratio'] = df_['ratio'].replace(np.nan,1).round(2)
    df_small.append(df_)

# df_small
df_ratios = pd.concat(df_small, ignore_index=True)
df_ratios.head(20)
```
